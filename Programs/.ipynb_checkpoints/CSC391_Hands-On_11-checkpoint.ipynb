## Hands-On 11: PWM_Two - Scoring using a PWM [Updated: March 13, 2023]

### Problem A: [5' Splice Sites]

From Hands-On Eleven, Question 2: <br><br>
2) To double-check the values you got in Table 2, run the program, score_9mers_MOG.py, that:
* Takes as input: 
 - <font color=blue>**mog_9mers_pwm.txt**</font> to read the values of the PWM
 - <font color=blue>**mog_9mers.txt**</font> to read the seven 9-mers 
* Computes the score of each sequence from <font color=blue>**mog_9mers.txt**</font>
* Writes each input sequence followed by its score into a file: <font color=blue>**mog_9mers_scores.txt**</font>.


The 9 mers of <font color=blue>**mog_9mers.txt**</font>:
* CAGGTAAGA 
* AAGGTGAGT 
* GAGGTACAG
* TAGGTGAGT
* TTGGTAAGT
* CAGGTGCAG
* TACGTAAGT

```python
# score_9mers_MOG.py
# Author: Sami Khuri
# Last updated: March 1, 2023
# Purpose: Program to score a sequence using the pwm generated by create_pwm.py
# Program uses the Python function open() and the Python methods read(), write(),
#    close() and split()
#

# Read a list of sequences of size 9, one per line, from the file mog_9mers.txt
input_file = open("mog_9mers.txt", "r")
input_data = input_file.readlines()
input_file.close()

# Create a list whose elements are the 9mers
input_list = []
for line in input_data:
    input_list.append(line.strip('\n'))
                      
output_file = open("mog_9mers_scores.txt","w")
output_file.close()

# Take as input mog_9mers_pwm.txt to read the values of the PWM
pwm = open("mog_9mers_pwm.txt","r")
pwm_data = pwm.read()
# split pwm_data and store the row as strings in the list "pwm_rows"
pwm_rows = pwm_data.split("\n")
# Create a list with the different values of the first row of "pwm_rows"
a = pwm_rows[0].split('\t')
# Remove the last empty element in a
a = a[0:9]
c = pwm_rows[1].split('\t')
c = c[0:9]
g = pwm_rows[2].split('\t')
g = g[0:9]
t = pwm_rows[3].split('\t')
t = t[0:9]
 
pwm.close()
output_file = open("mog_9mers_scores.txt", "w")
# Compute the score of each sequence read from the file mog_9mers.txt
for ele in input_list:
    total = 0
    for i in range(9):
        if ele[i] == 'A':
            total = total + float(a[i])
        elif ele[i] == 'C':
            total = total + float(c[i])
        elif ele[i] == 'G':
            total = total + float(g[i])
        elif ele[i] == 'T':
            total = total + float(t[i])
    # Write each input sequence followed by it's score into the file
    #     mog_9mers_scores.txt
    output_file.write(ele + ": " + str(round((total),3)) + "\n")

output_file.close()
            
```
